```python
import matplotlib.pyplot as plt
from matplotlib.cm import viridis

import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# models = ['naive\nbaseline', 'GPT-3\n(text-davinci-003)', 'FLAN-T5-Large\n(finetuned)', 'Expert\nPerformance']
# scores = [24.28, 50.45, 62.28, 72.04]
models = ['GPT-3\n(text-davinci-003)', 'FLAN-T5-Large\n(finetuned)', 'Expert\nPerformance']
scores = [50.45, 62.28, 72.04]

data = pd.DataFrame(data={
    'model': models,
    'performance': scores
})
```

```python
plt.rcParams['font.family'] = 'Helvetica Neue'
plt.rcParams['font.size'] = 16


fig, ax = plt.subplots(figsize=(10, 4))  # Set figure size
ax.grid(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
# ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

color=viridis(0.5 * data['performance'] / max(data['performance']))
data.plot(kind='bar', y='performance', x='model', ax=ax, color=color)
# plt.bar(data['model'], data['performance'], c='viridis')

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.1f'), (p.get_x() + p.get_width() / 2., p.get_height() + 1), ha = 'center', va = 'center', xytext = (0, 5), textcoords = 'offset points')


# Set the x-axis label
ax.set_xlabel('')

# Set the y-axis label
ax.set_ylabel('')

# Set the title of the plot
ax.set_title('Report-Extraction Performance (%F1)\n')

# Remove the legend
ax.get_legend().remove()

# Rotate x-axis ticks by 90 degrees
plt.xticks(rotation=00)
```
